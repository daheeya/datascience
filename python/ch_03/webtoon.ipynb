```python
import chromedriver_autoinstaller
chromedriver_autoinstaller.install()

from selenium import webdriver
browser=webdriver.Chrome()

from bs4 import BeautifulSoup
```

```python
url='https://comic.naver.com/webtoon?tab=mon'
browser.get(url)

html=browser.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup.find_all("div", {"class":"content"})
```

```python
webtoons=soup.select('ul.ContentList__content_list--q5KXY>li')
```

```python
len(webtoons)
```

```python
webtoon=webtoons[0]
title=webtoon.select('div>a')[0].text
title
author=webtoon.select('div>div>a.ContentAuthor__author--CTAAP')[0].text
author
rating=webtoon.select('div>span>span')[0].text.strip()
```

```python
rating
```

```python
webtoons=soup.select('ul.ContentList__content_list--q5KXY>li')
webtoon_data=[]
for webtoon in webtoons:
    title=webtoon.select('div>a')[0].text
    author=webtoon.select('div>a.ContentAuthor__author--CTAAP')[0].text
    rating=webtoon.select('div>span>span')[0].text.strip()
    webtoon_data.append([title,author,rating])
webtoon_data
```

```python
import time
```

```python
weeks=['mon','tue','wed','thu','fri','sat','sun']
webtoon_data=[]

for week in weeks:
    url=f'https://comic.naver.com/webtoon?tab={week}'
    browser.get(url)
    time.sleep(2)
    html=browser.page_source
    soup = BeautifulSoup(html, 'html.parser')

    webtoons=soup.select('ul.ContentList__content_list--q5KXY>li')
    for webtoon in webtoons:
        title=webtoon.select('div>a')[0].text
        author=webtoon.select('div>a.ContentAuthor__author--CTAAP')[0].text
        rating=webtoon.select('div>span>span')[0].text.strip()
        webtoon_data.append([week,title,author,rating])
webtoon_data
```

```python
import pandas as pd
```

```python
columns=['요일','제목','작가','평점']
df=pd.DataFrame(webtoon_data,columns=columns)
df
```
