```python
import time
```

```python
# title, category, subscriber, view, video
youtube=webdriver.Chrome()
channel_list=[]
for i in range(10):
    url=f'https://youtube-rank.com/board/bbs/board.php?bo_table=youtube&page={i+1}'
    youtube.get(url)
    time.sleep(2)

    html=youtube.page_source
    soup = BeautifulSoup(html, 'html.parser')

    channels=soup.select('form>table>tbody>tr')

    for channel in channels:
        rank=channel.select('td.rank')[0].text.strip(' \n')
        subscriber=channel.select('td.subscriber_cnt')[0].text
        category=channel.select('p.category')[0].text.strip(' \n')
        title=channel.select('h1>a')[0].text.strip(' \n \t')
        view=channel.select('td.view_cnt')[0].text
        video=channel.select('td.video_cnt')[0].text
        channel_list.append([rank,subscriber,category,title,view,video])
len(channel_list)
```

```python
df=pd.DataFrame(channel_list,columns=['rank','subscriber','category','title','view','video'])
df.tail()
```

```python
df.to_excel('../ch_02/files/youtube_rank.xlsx',index=False)
```

```python
df=df.astype({'rank':int})
df=df.sort_values(by='rank')
```
